```python
import numpy as np
import pandas as pd
```

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
tables = pd.read_html(URL)
df = tables[3] # Choose from the website
df.head()
```

```python
df.columns = range(df.shape[1]) # Convert columns name to numbers
df = df[[0,2]]
df = df.iloc[1:11,:] # Ten biggest economies in the world
df.columns = ['Country','GDP'] # Rename Columns
df.head()
```

```python
df['GDP'] = df['GDP'].astype(int) # Convert to integer
df['GDP'] = df['GDP']/1000 # Convert Million to Billion
df['GDP'] = np.round(df['GDP'],2) # Only two decimals
df.rename(columns = {'GDP':'GDP (Billion USD)'}, inplace = True)
df.head()
```

```python
df.to_csv('./Largest_economies.csv')
```
